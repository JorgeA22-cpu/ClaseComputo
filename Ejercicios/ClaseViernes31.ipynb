```julia
nmax = 100
c = 0.01
b = 150
f(x) = x*tan(x)-sqrt((b-x^2))

N_Eigenvalores = 10
X = zeros(N_Eigenvalores)

for i in 1:N_Eigenvalores
    k=0
    x0 = 0 + i
    x1 = 1 + i
    error0 = f(x0) 
    error1 = f(x1)
    error2 = min( abs(error0), abs(error1))
    tol = 0.00001
    while k <= nmax && abs(error2) > tol
        xr = x1 -( (x1-x0) * f(x1) ) / ( f(x1)-f(x0) )
        error2 = f(xr)
        x0 = x1 
        x1 = xr
        
        k = k+1
    end
    X[i] = xr
end
```
