```julia
using Plots
using SymPy
```

```julia
# Iteración de Muchas Condiciones Iniciales a la Vez 

iniciales = [x+y*im for x in -1:0.001:1, y in -1:0.001:1];
```

```julia
f12(x) = x - (x^4 + 4)/(4*x^3)
iterator = deepcopy(iniciales)
[iterator .= f12.(iterator) for i in 1:100] #One Liner
iterator
Etiqueta = deepcopy(iterator)
```

```julia
N = length(Etiqueta)
eps = 0.00001
function Etiquetador(Etiqueta)
    for i in 1:N 
        if abs(Etiqueta[i] - (-1-1im)) <eps
            Etiqueta[i] = 1
        elseif abs(Etiqueta[i] -(-1+1im)) < eps
            Etiqueta[i] = 2
        elseif abs(Etiqueta[i] - (1-1im)) < eps
            Etiqueta[i] = 3
        elseif abs(Etiqueta[i] - (1+1im)) < eps
            Etiqueta[i] = 4
        else
            Etiqueta[i] = 0
        end
    end
    return Etiqueta
end
Raiz = real(Etiquetador(Etiqueta))
```

```julia
X1 = real(iniciales)
Y1 = imag(iniciales)
heatmap(Raiz)#,c=:viridis)
```

## Ejercicio Para $x^4 - 4 = 0$

Las raíces son: $X_{1,2} = \pm \sqrt{2}$ y $X_{3,4} = \pm i\sqrt{2}$

```julia
f22(x) = x - (x^4 - 4)/(4*x^3)
iterator1 = deepcopy(iniciales)
[iterator1 .= f22.(iterator1) for i in 1:100] #One Liner
iterator1
Etiqueta1 = deepcopy(iterator1)
```

```julia
N1 = length(Etiqueta1)
eps1 = 0.0001
function Etiquetador2(Etiqueta1)
    for i in 1:N1 
        if abs(Etiqueta1[i] - (sqrt(2))) <eps1
            Etiqueta1[i] = 1
        elseif abs(Etiqueta1[i] -(-sqrt(2))) < eps1
            Etiqueta1[i] = 2
        elseif abs(Etiqueta1[i] - (sqrt(2)im)) < eps1
            Etiqueta1[i] = 3
        elseif abs(Etiqueta1[i] - (-sqrt(2)im)) < eps1
            Etiqueta1[i] = 4
        else
            Etiqueta1[i] = 0
        end
    end
    return Etiqueta1
end

Raiz2 = real(Etiquetador2(Etiqueta1))
```

```julia
heatmap(Raiz2)
```

```julia
@syms  z # Permitimos que z sea un símbolo
```

```julia
F = z^4+4 # Definimos la función F(z) = z^3 + 1
G = diff(F,z) # Derivamos F con respecto a z
H = z - F/G
```

```julia
Fx = solve(F) # Encontramos las raíces de F
V1 = lambdify.(Fx) #Convertir el vector en una función de Julia
V2 = [V1[i](1) for i in 1:4] # Al ser un valor constante, nos da igual realmente en qué valor evaluamos este vector, el resultado es el mismo. 
```

```julia
H1 = lambdify(H) # Convertimos H en una función de Julia
```

```julia

inicialesF = [x+y*im for x in -1:0.001:1, y in -1:0.001:1]; # Iteración de Muchas Condiciones Iniciales a la Vez 
nF = 50

iteratorF = deepcopy(inicialesF) # Copiamos las condiciones iniciales
[iteratorF .= H1.(iteratorF) for i in 1:nF] #One Liner, Aplica el método de Newton de n veces 
iteratorF
EtiquetaF = deepcopy(iteratorF)
```

```julia
NF = length(EtiquetaF)
epsF = 0.0001  # Tolerancia
function EtiquetadorF(EtiquetaF)
    for i in 1:NF
        if abs(EtiquetaF[i] - (V2[1])) <epsF
            EtiquetaF[i] = 1
        elseif abs(EtiquetaF[i] -(V2[2])) < epsF
            EtiquetaF[i] = 2
        elseif abs(EtiquetaF[i] - (V2[3])) < epsF
            EtiquetaF[i] = 3
        elseif abs(EtiquetaF[i] - (V2[4])) < epsF
            EtiquetaF[i] = 4
        else
            EtiquetaF[i] = 0
        end
    end
    return EtiquetaF
end

RaizF = real(EtiquetadorF(EtiquetaF))
```

```julia
heatmap(RaizF)
```

```julia
function EtiquetadorFPrueba(EtiquetaF)
    for i in 1:NF
        k = 1
        while abs(EtiquetaF[i]- (V2[k]) > epsF) || k > length(V2)
            k = k + 1
        end
        EtiquetaF[i] = k
    end
    return EtiquetaF
end
```

```julia
APrueba = reshape(1:16,4,4)
VPrueba = reverse(collect(1:16))
```

```julia
function Prueba(APrueba)
    for i in 1:length(APrueba)
        k = 1
        while abs(APrueba[i]- (VPrueba[k]) > epsF) || k > length(V2)
            k = k + 1
        end
        APrueba[i] = k
    end
    return EtiquetaF
end
```
